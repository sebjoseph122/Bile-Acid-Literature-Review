```python
import matplotlib.pyplot as plt # Version 3.7.3
import pandas as pd # Version 1.5.3
import seaborn as sns   # Version 0.13.2
import numpy as np  # Version 1.26.4
```

```python
peripheral_data = pd.read_csv("peripheral_data.csv")
peripheral_data
```

```python
column_names = peripheral_data.columns.to_list()
column_names
```

```python
ibas = column_names[8:23] #individual bas
ibas
```

```python
tbas = column_names[23:28]
tbas
```

```python
all_bas = ibas + tbas
all_bas
```

```python
ubas = ibas[0:5]
ubas
```

```python
#Dictionary with molecular weights
ba_mws = {"CA": 408.6, "CDCA": 392.6, "DCA": 392.6, "UDCA": 392.6, "LCA": 376.6, "G-CA":465.6, "G-CDCA":449.6, "G-DCA":449.6, "G-UDCA":449.6, "G-LCA":433.6,  "T-CA": 515.7, "T-CDCA": 499.7, "T-DCA":499.7, "T-UDCA":499.7, "T-LCA":483.7} #g/mol
```

```python
##Any non-numericals converted to NaN.
for ba in all_bas:
    peripheral_data[ba] = pd.to_numeric(peripheral_data[ba], errors='coerce')

peripheral_data
```

```python
#Converting Units to nM from uM
for ba in all_bas:
    peripheral_data.loc[peripheral_data['Units']== "uM", ba] *= 1000

# #Converting Units to nM from ug/ml
for ba in ibas:
    peripheral_data.loc[peripheral_data['Units'] == "ug/ml", ba]  *= (1000000/ba_mws[ba])

#Converting Units to nM from ng/ml
for ba in ibas:
    peripheral_data.loc[peripheral_data['Units'] == "ng/ml", ba] *= (1000/ba_mws[ba])

peripheral_data
```

```python
peripheral_data["Number of Patients"] = peripheral_data["Males"] + peripheral_data["Females"]
peripheral_data
```

```python
peripheral_data.to_csv("peripheral_data_nm.csv")
```

```python
#Dataframe in longform
x = []
y = []
z = []
a = []
b = []
c = []
d = []
e = []
f = []
g = []
h = []
i = []
j = []
for ba in ibas:
    x.extend([ba]*len(peripheral_data[ba]))
    y.extend(peripheral_data[ba].to_list())
    z.extend(peripheral_data["Fasted/Fed"].to_list())
    a.extend(peripheral_data["Serum/Plasma"].to_list())
    b.extend(peripheral_data["Analysis Technique"].to_list())
    c.extend(peripheral_data["Healthy/Unhealthy"].to_list())
    d.extend(peripheral_data["Exact Analysis Technique"].to_list())
    e.extend(peripheral_data["Starting Volume"].to_list())
    f.extend(peripheral_data["Average Age"].to_list())
    g.extend(peripheral_data["Number of Patients"].to_list())
    h.extend(peripheral_data["Protein Precipitation"])
    i.extend(peripheral_data["NAFLD/NASH/Cirrhosis"].to_list())
    j.extend(peripheral_data["Liver/Non-liver"].to_list())


DF = pd.DataFrame([x,y,z, a, b, c, d, e, f, g, h, i, j])
DF = DF.transpose()
DF = DF.rename(columns={0: "BA", 1:"Conc", 2:"Fasted/Fed", 3:"Serum/Plasma", 4:"Analysis Technique", 5:"Healthy/Unhealthy", 6:"Exact Analysis Technique", 7:"Starting Volume", 8:"Average Age", 9:"Number of Patients", 10:"Protein Precipitation", 11: "NAFLD/NASH/Cirrhosis", 12:"Liver/Non-Liver"})
```

```python
DF.to_csv("peripheral_ibas_longform.csv")
```
